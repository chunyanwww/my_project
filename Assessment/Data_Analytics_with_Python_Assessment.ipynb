## Job Role and Workload Level Impact on Mental Health

Instructions: Investigate the influence of job roles and workload level on the mental health.
- Create a new column `workload_level` that labels each entry as "High" if the `Hours_Worked_Per_Week` is above its mean, otherwise "Low." (5 marks)
- Group the dataset by `Industry` and calculate the average `Hours_Worked_Per_Week` for each combination. (5 marks)
- Use a bar chart to display the average `Stress_Level` for each job role, with separate bars for high and low stress levels. (5 marks)
- Analyze the results: Which job roles and workload levels appear to have the greatest impact on mental health? (5 marks)


```python
# Write code here
#Create a new column workload_level that labels each entry as "High" if the Hours_Worked_Per_Week is above its mean, otherwise "Low."


import numpy as np

#Use np.where for conditional assignment
df["Workload Level"] = np.where(
    df["Hours_Worked_Per_Week"] > df["Hours_Worked_Per_Week"].mean(),
    "High",
    "Low"
)
print("DataFrame with new column 'Workload Level':\n", df)

#Group the dataset by Industry and calculate the average Hours_Worked_Per_Week for each combination.

grouped_working_hours_by_industry = df_cleaned.groupby("Industry")["Hours_Worked_Per_Week"].mean()
print("\nAverage working hours per week in each industry:\n", grouped_working_hours_by_industry)


#Use a bar chart to display the average Stress_Level for each job role, with separate bars for high and low stress levels.

#Group the data by Job_Role and Stress_Level, then count the occurrences.
stress_level_counts = df.groupby(["Job_Role", "Stress_Level"]).size().unstack()

plt.figure(figsize=(6,8))
stress_level_counts.plot(kind="bar", edgecolor="black", colormap="coolwarm")
plt.title("Average stress level for each job role")
plt.xlabel("job role")
plt.ylabel("average stress level")
plt.show()


#Analyze the results: Which job roles and workload levels appear to have the greatest impact on mental health?
#Sales and Project Managers have immense pressure, with high expectations and demanding workloads.
#Designers, while facing creative stress, might experience relatively lower mental health impacts than Sales and PMs but still need to deal with tight deadlines.
#HR and Marketing roles seem to be better positioned in terms of mental health, with lower stress levels and better work-life balance.
#Long working hours alone are not the only factor contributing to poor mental health; the nature of the job, such as creativity in Design, customer interaction in Sales, team management in PM, plays a significant role.
```
